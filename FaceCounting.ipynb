```python
import os
import torch
from ultralytics import YOLO  # Assuming YOLO class is from the ultralytics library
import math
import pandas as pd
```

```python
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
```

```python
predict_path="/content/drive/MyDrive/test_annotated_images"
```

```python
img_paths = [os.path.join(predict_path, x) for x in os.listdir(predict_path)]
batch_size = 4
```

```python
model=YOLO("/content/drive/MyDrive/runs/detect/train2/weights/best.pt")
```

```python
results_data=[]
processed_count = 0
total_images = len(img_paths)
```

```python
def process_in_batches(img_paths, batch_size):
    global processed_count
    num_batches = math.ceil(len(img_paths) / batch_size)

    with torch.no_grad():
        for batch_idx in range(num_batches):
            # Select the current batch of images
            batch_imgs = img_paths[batch_idx * batch_size : (batch_idx + 1) * batch_size]

            # Run YOLO inference on the current batch
            results = model(batch_imgs, save=True, stream=True, batch=batch_size)

            # Process each result
            for img_path, result in zip(batch_imgs, results):
                # Count the number of faces (bounding boxes)
                head_count = len(result.boxes) if result.boxes else 0

                # Get the image name and add results to the list
                img_name = os.path.basename(img_path)
                results_data.append({"Name": img_name, "HeadCount": head_count})

                # Update and display processed count
                processed_count += 1
                print(f"Processed {processed_count}/{total_images} images")

                # Optional: display and save result
                #result.show()  # Display result on screen
               # result.save(filename=f"result_{img_name}")  # Save result image with unique filename

            # Clear CUDA cache after processing each batch to free up memory
            torch.cuda.empty_cache()
```

```python
process_in_batches(img_paths, batch_size)
```

```python
output_csv_path = "/content/drive/MyDrive/dataset/face_count_results.csv"
df = pd.DataFrame(results_data)
df.to_csv(output_csv_path, index=False)
print(f"Face count results saved to {output_csv_path}")
```
